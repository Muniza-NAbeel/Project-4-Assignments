```python
import qrcode      # Import `qrcode` for generating QR codes
import cv2           # Import `cv2` for QR code decoding
import numpy as np      # Import `numpy` for handling image data in decoding
from PIL import Image        # Import `PIL`'s `Image` for image processing in QR generation
from IPython.display import display      # Import `display` to show images in Jupyter Notebooks

def generate_qr(data, filename="qrcode.png"):
    qr = qrcode.QRCode(version=1, box_size=10, border=5)  # QR code version, box size, and border size

    qr.add_data(data)  # Adding data to the QR code

    qr.make(fit=True)  # Make the QR code fit the provided data

    img = qr.make_image(fill_color="purple", back_color="white")  # QR code with red foreground and white background

    img.save(filename)  # Saving the QR code image as a PNG file

    display(img)  # Displaying the image in Jupyter Notebooks

    print(f"✅ QR Code saved as {filename}")  # Confirmation that the QR code has been saved

def decode_qr(filename):
    img = cv2.imread(filename)  # Read the image using OpenCV

    detector = cv2.QRCodeDetector()  # Initialize the QR code detector

    data, _, _ = detector.detectAndDecode(img)  # Attempt to decode the QR code in the image

    if data:
        print(f"✅ Decoded Data: {data}")  # Output the decoded data from the QR code
    else:
        print("❌ No QR code found! Try using a white background.")  # Inform that no QR code was detected

generate_qr("Successfully Generated", "my_qr.png")  # Generate the QR code and save it

decode_qr("my_qr.png")  # Attempt to decode the generated QR code
```
